```python
import turtle
import math

# Set up the turtle screen
screen = turtle.Screen()
screen.bgcolor("white")

# Create a turtle instance
t = turtle.Turtle()
t.speed(1)  # Set the drawing speed (1 is slowest, 10 is fastest)
t.pensize(2)  # Set the pen size

# Define a function to draw a rectangle
def draw_rectangle(x, y, width, height, color):
    t.penup()
    t.goto(x, y)
    t.pendown()
    t.color(color)
    for _ in range(2):
        t.forward(width)
        t.left(90)
        t.forward(height)
        t.left(90)

# Define a function to draw a circle
def draw_circle(x, y, radius, color):
    t.penup()
    t.goto(x, y - radius)
    t.pendown()
    t.color(color)
    t.circle(radius)

# Define a function to translate a 2D object
def translate(dx, dy):
    t.penup()
    x, y = t.pos()
    t.goto(x + dx, y + dy)
    t.pendown()

# Define a function to rotate a 2D object
def rotate(angle):
    t.penup()
    x, y = t.pos()
    t.goto(x, y)
    t.setheading(angle)
    t.pendown()

# Define a function to scale a 2D object
def scale(sx, sy):
    t.penup()
    x, y = t.pos()
    t.goto(x * sx, y * sy)
    t.pendown()

# Draw a rectangle
draw_rectangle(-200, 0, 100, 50, "blue")

# Translate the rectangle
translate(200, 0)
draw_rectangle(-200, 0, 100, 50, "blue")

# Rotate the rectangle
rotate(45)
draw_rectangle(-200, 0, 100, 50, "blue")

# Scale the rectangle
scale(2, 2)
draw_rectangle(-200, 0, 100, 50, "blue")

# Draw a circle
draw_circle(100, 100, 50, "red")

# Translate the circle
translate(200, 0)
draw_circle(100, 100, 50, "red")

# Rotate the circle (Note: This operation does not visually change a circle)
rotate(45)
draw_circle(100, 100, 50, "red")

# Scale the circle
scale(2, 2)
draw_circle(100, 100, 50, "red")

# Keep the window open until it's closed
turtle.done()
```
