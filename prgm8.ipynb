```python

import cv2
import numpy as np

# Load an image from file
# Replace 'path_to_your_image.jpg' with the actual path to your image file
image_path = 'C:\\Users\\cheta\\OneDrive\\Desktop\\CG&IP prgm\\demoimg.jpg'
image = cv2.imread(image_path)

# Function to rotate the image by a specific angle
def rotate_image(image, angle):
    # Get the dimensions of the image
    (h, w) = image.shape[:2]
    # Compute the center of the image
    center = (w // 2, h // 2)
    # Get the rotation matrix for the given angle
    matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    # Apply the rotation to the image
    rotated_image = cv2.warpAffine(image, matrix, (w, h))
    return rotated_image

# Function to scale the image by a specific factor
def scale_image(image, scale_factor):
    # Get the dimensions of the image
    (h, w) = image.shape[:2]
    # Compute the new size of the image
    new_size = (int(w * scale_factor), int(h * scale_factor))
    # Resize the image
    scaled_image = cv2.resize(image, new_size, interpolation=cv2.INTER_LINEAR)
    return scaled_image

# Function to translate the image by a specific offset in x and y directions
def translate_image(image, tx, ty):
    # Get the dimensions of the image
    (h, w) = image.shape[:2]
    # Create a translation matrix with the given x and y offsets
    matrix = np.float32([[1, 0, tx], [0, 1, ty]])
    # Apply the translation to the image
    translated_image = cv2.warpAffine(image, matrix, (w, h))
    return translated_image

# Apply transformations to the image
# Rotate the image by 45 degrees
rotated_image = rotate_image(image, 45)
# Scale the image by 1.5 times its original size
scaled_image = scale_image(image, 1.5)
# Translate the image by 100 pixels right and 50 pixels down
translated_image = translate_image(image, 100, 50)

# Display the original and transformed images
# Use OpenCV's imshow to show the images in separate windows
cv2.imshow('Original Image', image)
cv2.imshow('Rotated Image', rotated_image)
cv2.imshow('Scaled Image', scaled_image)
cv2.imshow('Translated Image', translated_image)

# Wait indefinitely until a key is pressed
cv2.waitKey(0)
# Close all OpenCV windows
cv2.destroyAllWindows()
```
